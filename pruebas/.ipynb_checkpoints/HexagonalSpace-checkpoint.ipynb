HEXAGONAL SPACE

```julia
using Agents
import Agents: add_agent_to_space!, remove_agent_from_space!, 
              random_position, nearby_ids, nearby_positions

# Estructura del Espacio
struct HexagonalSpaceSingle{P} <: Agents.AbstractGridSpace{2, P}
    s::Matrix{Int}      # Matriz de IDs (0 si está vacía)
    dims::NTuple{2, Int}
    periodic::Bool
end

# Constructor
function HexagonalSpaceSingle(dims::NTuple{2, Int}; periodic = false)
    s = zeros(Int, dims)
    return HexagonalSpaceSingle{periodic}(s, dims, periodic)
end
```

```julia
function mod_pos(pos, dims)
    return (mod1(pos[1], dims[1]), mod1(pos[2], dims[2]))
end

# 1. nearby_positions: Devuelve las coordenadas de las celdas cercanas
function nearby_positions(pos::NTuple{2, Int}, model::ABM{<:HexagonalSpaceSingle}, r::Real = 1)
    dims = model.space.dims
    periodic = model.space.periodic
    res = NTuple{2, Int}[]
    
    ri = Int(floor(r))
    # Iteramos en el "bounding box" del radio r
    for dq in -ri:ri
        for dr in max(-ri, -dq-ri):min(ri, -dq+ri)
            dq == dr == 0 && continue
            new_pos = (pos[1] + dq, pos[2] + dr)
            
            if periodic
                push!(res, mod_pos(new_pos, dims))
            else
                if 1 <= new_pos[1] <= dims[1] && 1 <= new_pos[2] <= dims[2]
                    push!(res, new_pos)
                end
            end
        end
    end
    return unique(res)
end

# 2. nearby_ids: Aprovecha nearby_positions para sacar los IDs
function Agents.nearby_ids(pos::NTuple{2, Int}, model::ABM{<:HexagonalSpaceSingle}, r::Real = 1)
    # Aquí llamamos a nuestra función de posiciones
    positions = nearby_positions(pos, model, r)
    ids = Int[]
    space = abmspace(model) # <--- API Correcta
    for p in positions
        id = space.s[p...]
        id != 0 && push!(ids, id)
    end
    return ids
end

Agents.positions(space::HexagonalSpaceSingle) = CartesianIndices(space.s)

# 2. También necesitamos decirle cómo leer los IDs en una posición específica 
# (como es Single, o devuelve el ID o nada)
function Agents.ids_in_position(pos::NTuple{2, Int}, model::ABM{<:HexagonalSpaceSingle})
    id = abmspace(model).s[pos...]
    return id == 0 ? () : (id,)
end

# 3. Y para que funciones como 'fill_space!' funcionen:
Agents.has_empty_positions(model::ABM{<:HexagonalSpaceSingle}) = any(id -> id == 0, abmspace(model).s)
```

```julia
function create_space()
    stype = "hexagonal"
    dims = [10, 10]
    periodic = false

    if stype == "grid"
        return HexagonalSpaceSingle(dims; periodic = periodic)
    end
end
```

```julia
# Para meter al agente en la matriz interna
function Agents.add_agent_to_space!(agent, model::ABM{<:HexagonalSpaceSingle})
    # Usamos abmspace(model) en lugar de model.space
    space = abmspace(model)
    pos = agent.pos
    if space.s[pos...] != 0
        error("La celda hexagonal $(pos) ya está ocupada.")
    end
    space.s[pos...] = agent.id
    return agent
end

# Para sacar al agente (limpiar la celda)
function Agents.remove_agent_from_space!(agent, model::ABM{<:HexagonalSpaceSingle})
    abmspace(model).s[agent.pos...] = 0
    return agent
end

# Extra: Mover agentes es automático, pero por eficiencia puedes definirlo:
function Agents.move_agent!(agent, pos::NTuple{2, Int}, model::ABM{<:HexagonalSpaceSingle})
    # Quitamos de la posición vieja
    model.space.s[agent.pos...] = 0
    # Ponemos en la nueva
    agent.pos = pos
    model.space.s[pos...] = agent.id
    return agent
end
```

```julia
import CairoMakie
using Agents
using CairoMakie
function Agents.abmplot_pos(agent, model::ABM{<:HexagonalSpaceSingle})
    q, r = agent.pos
    x = sqrt(3) * (q + r/2.0)
    y = 1.5 * r
    
    return (Float32(x), Float32(y))
end
```

```julia
@agent struct Cell(GridAgent{2})
    status::Bool # true = Vivo, false = Muerto
end
```

```julia
function hex_life_step!(model)
    next_states = Dict{Int, Bool}()
    
    for agent in allagents(model)
        neighbor_ids = nearby_ids(agent.pos, model, 1)
        living_neighbors = count(id -> model[id].status, neighbor_ids)
        
        if agent.status
            next_states[agent.id] = (living_neighbors == 2 || living_neighbors == 3)
        else
            next_states[agent.id] = (living_neighbors == 2)
        end
    end
    
    # 2. Aplicamos los cambios
    for (id, new_status) in next_states
        model[id].status = new_status
    end
end
```

```julia
function initialize_model(; dims=(30, 30), density=0.3)
    # 1. Crear el espacio
    space = HexagonalSpaceSingle(dims; periodic=true)
    
    # 2. Crear el modelo
    model = StandardABM(Cell, space; model_step! = hex_life_step!)
    
    # 3. Llenar la rejilla usando el iterador de posiciones que ya arreglamos
    for pos in positions(model)
        if rand() < density
            # En un SpaceSingle, add_agent! mete el agente en esa celda específica
            add_agent!(pos, model, true) 
        else
            # Aunque esté muerta, en el Juego de la Vida necesitamos el agente
            # para que tenga un estado .status = false
            add_agent!(pos, model, false)
        end
    end
    
    return model
end

```

```julia
function agent_color(a)
    return a.status ? :black : :white
end

const HEX_SHAPE = Makie.Polygon(Point2f[
    (cos(a), sin(a)) for a in range(0, 2π, length=7)[1:6]
])

function save_hex_video(model, filename="hex_life.mp4")
    w, h = model.space.dims
    limit_x = (0.0, sqrt(3) * (w + h/2.0))
    limit_y = (0.0, 1.5 * h)

    abmvideo(
        filename, model;
        agent_color = agent_color,
        agent_marker = HEX_SHAPE,
        agent_size = 15,
        frames = 100,
        framerate = 10,
        title = "Hexagonal Game of Life",
        axis = (limits = (limit_x, limit_y), aspect = DataAspect())
    )
end
```

```julia
model = initialize_model(dims=(40, 40), density=0.2)
```

```julia
save_hex_video(model, "mi_juego_hexagonal.mp4")
```
